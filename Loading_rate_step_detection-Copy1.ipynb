```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
folder_name = Path(r'')
filename = ""
```

```python
rawdata = np.loadtxt(folder_name/filename, skiprows=2)
C = 182
```

```python
rawTime = rawdata[:,0]
rawMagPos = rawdata[:,1]
rawForce = C*(np.exp(-(13.5-rawMagPos)/0.695)+0.138*np.exp(-(13.5-rawMagPos)/1.59))
rawExt = rawdata[:,2]
#MagPos_cutoff = 8.0
MagPos_cutoff = min(rawMagPos)
print(MagPos_cutoff)
```

```python
cycleflag = 0
newTime = [[]]
newExt = [[]]
Force = [[]]
datalength = len(rawTime)
for i in range(datalength-1):
  if np.isnan(rawExt[i]):
    rawExt[i]=rawExt[i-1]
  if rawMagPos[i]>MagPos_cutoff:
    if all(rawMagPos[i+1:i+10]<=MagPos_cutoff) :
      cycleflag = cycleflag + 1
      newTime.append([])
      newExt.append([])
      Force.append([])
    else:
      newTime[cycleflag].append(rawTime[i])
      newExt[cycleflag].append(rawExt[i])
      Force[cycleflag].append(rawForce[i])
```

```python
from scipy import fftpack
fig = plt.figure(figsize=[10,8])
plot_cycle_num = np.array([0,1,2])
colors = ['b','c','g','m','r','y']
for n in plot_cycle_num:
  max_index = np.argmax(Force[n])
  #plt.plot(Force[n][5:max_index], newExt[n][5:max_index], color='lightgrey', linewidth=1, zorder=0)
  #plt.plot(Force[n][5:max_index], newExt[n][5:max_index], color='lightgrey', linewidth=1, label='Original signal', zorder=0)
  #plt.plot(Force[n][max_index:-5], newExt[n][max_index:-5], color='lightgrey',linewidth=2, label='Original signal R', zorder=1)
cycle = 1
for m in plot_cycle_num:
  l_rate = [0.1]
  #l_rate = [10, 1, 0.1]
  #temp = [23, 37]
  color_index = np.where(plot_cycle_num == m)
  sig_fft = fftpack.fft(newExt[m])
  sample_freq = fftpack.fftfreq(len(newExt[m]), d=0.005)
  high_freq_fft = sig_fft.copy()
  high_freq_fft[np.abs(sample_freq) > 10] = 0
  filtered_sig = np.real(fftpack.ifft(high_freq_fft))
  max_index = np.argmax(Force[m])
  plt.plot(Force[m][5:max_index], filtered_sig[5:max_index], color=colors[color_index[0][0]],linewidth=2, label='cycle %.d'%cycle, zorder=2)
  #plt.plot(Force[m][max_index:-5], filtered_sig[max_index:-5], color='grey',linewidth=2, label='Filtered signal R', zorder=1)
  cycle += 1
#plt.ylim(-40,421)
#plt.xlim(0,80)
#plt.yticks(np.arange(-40, 421, step=40))
#plt.xticks(np.arange(0, 81, step=10))
plt.ylabel('Bead height(nm)', size=36)
plt.xlabel('Force(pN)', size=36)
plt.title('FLNC R9-15 1pN/s', size=36)
plt.tick_params(labelsize=36)
plt.legend(loc='upper left',framealpha=1, prop = {'size':20})
#plt.savefig(figure_path/'filaminCmonomer2pns_curves.eps', format='eps', bbox_inches = 'tight')
plt.show()
```
